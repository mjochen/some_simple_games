```python
def get_punten(lijn):
    lijn.sort()
    # Constants Declaration
    n = 3
    prev = -1
    count = 0
    # print(lijn)

    # Iterating
    amounts = []
    for item in lijn:
        if item == prev:
            count = count + 1
        else:
            if count:
                amounts.append(count)
            count = 1

        prev = item
    amounts.append(count)

    amounts.sort()

    if amounts[-1] == 5:
        return 10
    if amounts[-1] == 4:
        return 6
    if amounts[-1] == 3:
        if amounts[-2] == 2:
            return 8
        return 3
    if amounts[-1] == 2:
        if amounts[-2] == 2:
            return 3
        return 1

    for i in range(len(lijn)-1):
        if lijn[i]+1 != lijn[i+1]:
            return 0
    if 7 in lijn:
        return 8
    return 12


def get_alle_punten(vakje):
    totaal = 0
    for lijn in vakje:
        totaal += get_punten(lijn.copy())

    for i in range(len(vakje[0])):
        # print(i)
        # print([l[i] for l in vakje])
        totaal += get_punten([l[i] for l in vakje])
        # print(totaal)

    # print([vakje[i][i] for i in range(len(vakje))])
    totaal += get_punten([vakje[i][i] for i in range(len(vakje))]) *2
    # print(totaal)
    totaal += get_punten([vakje[i][-1-i] for i in range(len(vakje))]) *2
    # print(totaal)
    return totaal
```

```python
print(get_punten([4,9,8,3,4]))
print(get_punten([8,8,7,7,8]))
```

```python
vakje = [[4,9,8,3,4], [10,4,7,3,10], [9,6,8,3,6], [8,5,7,4,6], [6,12,7,4,5]]

get_alle_punten(vakje)
```
